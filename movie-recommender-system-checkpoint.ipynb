```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# i have to convert this format into ['action' , 'adventure' , 'fantasy' .scifi']
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')




def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L  # Move this back so it aligns with the 'for' statement
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# same think we will do for keywords column as well.
movies['keywords'] =movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
# for cast column i have only need the actor real name. not their movie character name.
# and only take top 3 actor name.


def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# in movie crew i only need  director name
    
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L  # This must be aligned with the 'for'
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# the overveiw column is a string  so we convert it into list.
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# before merging these cell .we have to perform a transformation to delete the space between the names . because some names have same first word .it might conjused over model.
# 'sam worthington'
# 'samworthington'
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
# now we a have a new column name tags . which summarise main column . so now we remove these columns.
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
# here we convert tags data into string.
```

```python
# TO FIX THE PROBLEM OF SAME MEANING IN WORDS (102 LINE OF CODE)
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```
